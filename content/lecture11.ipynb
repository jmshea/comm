```python
import numpy as np
import scipy as sp
import scipy.signal

import matplotlib.pyplot as plt
%matplotlib inline
```

## Triangular Message Signal

```python
# Triangle wave generation (scipy only seems to have sawtooth)
def triangle_wave(t, period, amplitude=1, offset=0):
    m = [-4*(np.abs(-(np.fmod(t[i]*(amplitude)/period, amplitude)-amplitude/2)))+amplitude +offset for i in range(len(t))]
    m = np.array(m)
    return m;
```

```python
tri_amplitude = 1
tri_period = 2e-4

max_time = 1*tri_period
t_steps = 20000
t = np.linspace(0,max_time,t_steps)

m = triangle_wave(t, tri_period, tri_amplitude, tri_amplitude)


plt.plot(t*1e6,m);
```

```python
def square_wave(t, period, amplitude=1, offset=0):

    m = amplitude*sp.signal.square(t*(2*np.pi/period))+ offset

    return m
```

```python
amplitude = 1
period = 2e-4

max_time = 4*period
t_steps = 20000
t = np.linspace(0,max_time,t_steps)

m = square_wave(t, period, amplitude, amplitude)


plt.plot(t*1e6,m);
```
